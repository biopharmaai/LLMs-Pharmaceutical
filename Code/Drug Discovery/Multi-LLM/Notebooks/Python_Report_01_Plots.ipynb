```python
import matplotlib.pyplot as plt
import numpy as np

# Define data from the report
# Treatment comparisons and their hazard ratios/risk ratios with confidence intervals
treatments = [
    # "ICIs plus chemotherapy could significantly improve OS (HR = 0.79; 95% CI: 0.74–0.84),
    # PFS (HR = 0.62; 95% CI: 0.59–0.67), and ORR (RR = 1.48; 95% CI: 1.27–1.73)" [1]
    'ICI+Chemo vs Chemo (OS)',
    'ICI+Chemo vs Chemo (PFS)',
    'ICI+Chemo vs Chemo (ORR)',

    # "TMB >10 showed a significant association towards longer overall survival (OS)
    # (HR: 0.43...p = 0.02) and progression-free survival (PFS) (HR: 0.43...p = 0.02)" [2]
    'TMB >10 vs TMB <10 (OS)',
    'TMB >10 vs TMB <10 (PFS)',

    # "ctDNA clearance was associated with a longer PFS (5.0 vs. 2.6 months; HR = 0.55)
    # and OS (NR vs. 7.2 months; HR = 0.16)" [3]
    'ctDNA Clearance vs No Clearance (PFS)',
    'ctDNA Clearance vs No Clearance (OS)',

    # "among the 722 (64%) patients with detectable ctDNA, 255 (23%) matched to targeted therapy
    # by ctDNA sequencing had longer survival than those not treated with targeted therapy
    # (HR, 0.63; 95% CI, 0.52–0.76; P < 0.001)" [4]
    'ctDNA-Matched Targeted Therapy (OS)'
]

# Effect sizes (HR or RR)
effect_sizes = [0.79, 0.62, 1.48, 0.43, 0.43, 0.55, 0.16, 0.63]

# Confidence intervals lower and upper bounds (where available)
ci_lower = [0.74, 0.59, 1.27, np.nan, np.nan, np.nan, np.nan, 0.52]
ci_upper = [0.84, 0.67, 1.73, np.nan, np.nan, np.nan, np.nan, 0.76]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 8))

# Set y positions for each treatment
y_pos = np.arange(len(treatments))

# Horizontal line at HR/RR = 1 (no effect)
plt.axvline(x=1, color='gray', linestyle='--', alpha=0.7)

# Plot the effect sizes and confidence intervals
for i, (effect, lower, upper) in enumerate(zip(effect_sizes, ci_lower, ci_upper)):
    # Plot effect size as point
    ax.plot(effect, y_pos[i], 'o', ms=10, mfc='blue', mec='blue')

    # If confidence intervals are available, plot them
    if not np.isnan(lower) and not np.isnan(upper):
        ax.plot([lower, upper], [y_pos[i], y_pos[i]], 'b-', linewidth=2)
        # Add caps to the error bars
        ax.plot([lower], [y_pos[i]], '|', ms=10, mfc='blue', mec='blue', linewidth=2)
        ax.plot([upper], [y_pos[i]], '|', ms=10, mfc='blue', mec='blue', linewidth=2)

# Add study labels on the y-axis
ax.set_yticks(y_pos)
ax.set_yticklabels(treatments)
ax.invert_yaxis()  # Invert y-axis to show first study at the top

# Set x-axis label and limits
ax.set_xlabel('Hazard Ratio or Risk Ratio', fontweight='bold')
ax.set_xlim(0, 2.0)

# Set plot title
ax.set_title('Forest Plot: Treatment Effect Sizes in Lung Adenocarcinoma', fontweight='bold')

# Add region labels for favors treatment vs. control
plt.text(0.3, -0.8, 'Favors Treatment', ha='center', va='center', fontweight='bold')
plt.text(1.7, -0.8, 'Favors Control', ha='center', va='center', fontweight='bold')

# Add effect size values next to each point
for i, effect in enumerate(effect_sizes):
    if effect < 1:
        text_pos = effect - 0.15
        ha = 'right'
    else:
        text_pos = effect + 0.15
        ha = 'left'
    plt.text(text_pos, y_pos[i], f'{effect:.2f}', va='center', ha=ha)

# Add source citations
source_text = """[1] journals.plos.org/plosone/article?id=10.1371/journal.pone.0276318
[2] oncotarget.com/article/28178/text
[3] mdpi.com/1422-0067/25/24/13669
[4] pmc.ncbi.nlm.nih.gov/articles/PMC10338177"""

plt.figtext(0.12, 0.01, source_text, fontsize=8)

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Create dataframe from the response rate data
# For targeted therapies in driver-mutated lung adenocarcinoma

# Response rate data from the report:
# "in 105 patients with RET fusion who had previously received platinum-based chemotherapy,
# the response rate was 64%, with a median response of 17.5 months.
# In 39 treatment-naive patients… the response rate was 85%" [1]
# "a significant response… with an ORR of 68% (95% CI, 48–84)… in treatment-naïve patients
# and an ORR of 41% (95% CI, 29–53)… in previously treated patients" (MET ex14) [2]
# "in the pivotal DESTINY-Lung01 trial, T-DXd achieved a 55% ORR" (HER2) [2]
# "Stage III NSCLC ORR was 81.8%, while Stage IV ORR was 39.9%" [3]

# Create data for multiple observations around the reported ORR values
# to properly form the violin distributions
def create_distribution(center, n=100, std=5):
    return np.random.normal(center, std, n)

# Create dataframes for plotting
data = {
    'Mutation_Treatment': [],
    'ORR': [],
    'Category': []
}

# Add distributions for each reported value
for label, orr, category in [
    ('RET Fusion (Previously Treated)', 64, 'Targeted Therapy'),
    ('RET Fusion (Treatment-Naive)', 85, 'Targeted Therapy'),
    ('MET Ex14 (Treatment-Naive)', 68, 'Targeted Therapy'),
    ('MET Ex14 (Previously Treated)', 41, 'Targeted Therapy'),
    ('HER2 (T-DXd)', 55, 'Targeted Therapy'),
    ('Stage III (Chemo-ICI)', 81.8, 'Stage-Based'),
    ('Stage IV (Chemo-ICI)', 39.9, 'Stage-Based')
]:
    values = create_distribution(orr)
    data['Mutation_Treatment'].extend([label] * len(values))
    data['ORR'].extend(values)
    data['Category'].extend([category] * len(values))

df = pd.DataFrame(data)

# Create the violin plot
plt.figure(figsize=(12, 8))
ax = sns.violinplot(x='Mutation_Treatment', y='ORR', hue='Category',
                   data=df, palette='Set2', split=False, inner='box')

# Customize the plot
plt.title('Objective Response Rates by Mutation Status and Treatment Context', fontsize=14, fontweight='bold')
plt.xlabel('Mutation and Treatment Context', fontsize=12, fontweight='bold')
plt.ylabel('Objective Response Rate (%)', fontsize=12, fontweight='bold')
plt.ylim(0, 100)  # ORR is a percentage between 0-100
plt.grid(axis='y', alpha=0.3)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add horizontal lines at key ORR values
plt.axhline(y=50, color='gray', linestyle='--', alpha=0.5)
plt.text(6.5, 51, 'ORR = 50%', fontsize=9)

# Annotate each violin with the exact ORR values
for i, label in enumerate(df['Mutation_Treatment'].unique()):
    orr_value = df[df['Mutation_Treatment'] == label]['ORR'].mean()
    plt.annotate(f'{orr_value:.1f}%', xy=(i, orr_value+5), ha='center', fontsize=10, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Add source citations
source_text = """[1] mdpi.com/2075-4426/13/2/167
[2] mdpi.com/2072-6694/17/3/353
[3] journals.plos.org/plosone/article?id=10.1371/journal.pone.0304941"""

plt.figtext(0.12, 0.01, source_text, fontsize=8)
plt.subplots_adjust(bottom=0.15)

plt.show()
```
