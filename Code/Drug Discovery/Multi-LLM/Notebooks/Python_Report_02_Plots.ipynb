```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
from lifelines import KaplanMeierFitter

# Data from ALK-positive NSCLC study comparing lorlatinib versus crizotinib
# "the 5-year PFS rate was 60% with lorlatinib versus 8% with crizotinib" [1]
# "median PFS not reached for lorlatinib versus 9.3 months... for crizotinib (HR 0.28... p < 0.001)" [1]
# [1] mdpi.com/2072-6694/17/3/353

# Set up time points in months (0-60 months = 5 years)
months = np.arange(0, 61, 1)

# Simulated data based on the reported outcomes
# Crizotinib: median PFS 9.3 months, 5-year PFS 8%
# Lorlatinib: median PFS not reached, 5-year PFS 60%

# Create survival probabilities
def generate_survival_curve(median_survival, survival_at_5years):
    surv = np.ones(61)

    # Set 5-year point
    surv[60] = survival_at_5years

    # If median is known, set survival at 0.5 at that time point
    if median_survival is not None and median_survival <= 60:
        surv[median_survival] = 0.5

    # Generate curve (approximation)
    if median_survival is not None and median_survival <= 60:
        # Before median
        for i in range(1, median_survival):
            surv[i] = 1 - ((0.5 / median_survival) * i)
        # After median to 5 years
        for i in range(median_survival + 1, 61):
            surv[i] = 0.5 * (survival_at_5years/0.5) ** ((i - median_survival)/(60 - median_survival))
    else:
        # Median not reached, use 5-year survival to estimate curve
        for i in range(1, 61):
            surv[i] = 1 - ((1 - survival_at_5years) * (i/60) ** 1.5)

    return surv

# Generate survival data
crizotinib_surv = generate_survival_curve(9, 0.08)
lorlatinib_surv = generate_survival_curve(None, 0.60)

# Create plot
plt.figure(figsize=(10, 6))

# Plot survival curves
plt.step(months, crizotinib_surv, where='post', color='red', linewidth=2)
plt.step(months, lorlatinib_surv, where='post', color='blue', linewidth=2)

# Add markers at key points
plt.scatter(9, 0.5, color='red', s=80, zorder=5)  # Median PFS for crizotinib
plt.scatter(60, 0.08, color='red', s=80, zorder=5)  # 5-year PFS for crizotinib
plt.scatter(60, 0.60, color='blue', s=80, zorder=5)  # 5-year PFS for lorlatinib

# Formatting
plt.xlabel('Time (Months)', fontsize=12)
plt.ylabel('Progression-Free Survival Probability', fontsize=12)
plt.title('Kaplan-Meier Curve: Progression-Free Survival in ALK+ Lung Adenocarcinoma', fontsize=14)
plt.ylim(0, 1.05)
plt.xlim(0, 60)
plt.grid(True, linestyle='--', alpha=0.7)

# Legend
legend_elements = [
    Line2D([0], [0], color='blue', lw=2, label='Lorlatinib (5-yr PFS: 60%)'),
    Line2D([0], [0], color='red', lw=2, label='Crizotinib (5-yr PFS: 8%, median PFS: 9.3 months)')
]
plt.legend(handles=legend_elements, loc='lower left', fontsize=10)

plt.text(62, 0.5, 'HR = 0.28, p < 0.001', fontsize=10)
plt.text(0, -0.1, 'Citation: [1] mdpi.com/2072-6694/17/3/353', fontsize=8, transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from meta-analysis of ICIs plus chemotherapy versus chemotherapy alone
# "ICIs plus chemotherapy could significantly improve OS (HR = 0.79; 95% CI: 0.74–0.84),
# PFS (HR = 0.62; 95% CI: 0.59–0.67), and ORR (RR = 1.48; 95% CI: 1.27–1.73)
# when compared to chemotherapy" [1]
# [1] journals.plos.org/plosone/article?id=10.1371/journal.pone.0276318

# Data from TMB study
# "TMB >10 showed a significant association towards longer overall survival (OS)
# (HR: 0.43...p = 0.02) and progression-free survival (PFS) (HR: 0.43...p = 0.02)
# in patients treated with first-line immunotherapy" [2]
# [2] oncotarget.com/article/28178/text

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Define data
outcomes = [
    'ICIs + Chemo: OS',
    'ICIs + Chemo: PFS',
    'ICIs + Chemo: ORR',
    'TMB >10: OS',
    'TMB >10: PFS'
]

# HR values and confidence intervals
hrs = [0.79, 0.62, 1.48, 0.43, 0.43]
lower_cis = [0.74, 0.59, 1.27, None, None]  # None where CI not provided
upper_cis = [0.84, 0.67, 1.73, None, None]

# Calculate error bars where available
lower_errors = [hrs[i] - lower_cis[i] if lower_cis[i] is not None else 0 for i in range(len(hrs))]
upper_errors = [upper_cis[i] - hrs[i] if upper_cis[i] is not None else 0 for i in range(len(hrs))]

# Y positions for the outcomes
y_pos = np.arange(len(outcomes))

# Create horizontal error bars
ax.errorbar(
    hrs[:3], y_pos[:3],
    xerr=[lower_errors[:3], upper_errors[:3]],
    fmt='o',
    capsize=5,
    color='blue',
    ecolor='blue',
    markersize=8
)

ax.errorbar(
    hrs[3:], y_pos[3:],
    xerr=None,  # No CI available
    fmt='o',
    capsize=5,
    color='green',
    markersize=8
)

# Add vertical line at HR = 1.0 (no effect)
ax.axvline(x=1.0, color='black', linestyle='--', alpha=0.7)

# Add p-values where available
plt.text(hrs[3] - 0.05, y_pos[3] + 0.15, 'p = 0.02', fontsize=9)
plt.text(hrs[4] - 0.05, y_pos[4] + 0.15, 'p = 0.02', fontsize=9)

# Set labels and title
ax.set_yticks(y_pos)
ax.set_yticklabels(outcomes)
ax.set_xlabel('Hazard Ratio / Risk Ratio', fontsize=12)
ax.set_title('Forest Plot: Treatment Effects in Lung Adenocarcinoma', fontsize=14)

# Add interpretation guide
ax.text(0.3, -0.3, 'Favors Treatment', fontsize=10, ha='center')
ax.text(1.7, -0.3, 'Favors Control', fontsize=10, ha='center')

# Add citations
ax.text(0.01, -0.2, 'Citations: [1] journals.plos.org/plosone/article?id=10.1371/journal.pone.0276318',
       transform=ax.transAxes, fontsize=8)
ax.text(0.01, -0.25, '              [2] oncotarget.com/article/28178/text',
       transform=ax.transAxes, fontsize=8)

# Set x-axis limits
ax.set_xlim(0.3, 2.0)

# Add grid lines
ax.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats

# Data from metastatic site study
# "the median survival times of patients with single lung metastasis (11 months; 95% CI 0.470–0.516)
# was better than brain metastasis (5 months; 95% CI 0.465–0.508), bone metastasis (5 months;
# 95% CI 0.504–0.547) and liver metastasis (6 months; 95% CI 0.483–0.590)" [1]
# [1] pmc.ncbi.nlm.nih.gov/articles/PMC10082283

# Define metastatic sites and their median survival times (in months)
sites = ['Lung', 'Brain', 'Bone', 'Liver']
median_survival = [11, 5, 5, 6]

# Generate simulated distributions based on reported medians
# Using log-normal distributions which are common for survival data
np.random.seed(42)  # For reproducibility

# Create simulated data
def generate_simulated_data(median, size=500):
    # For log-normal distribution, median = exp(mu)
    mu = np.log(median)
    # Assuming a reasonable sigma value to create variability
    sigma = 0.5
    return np.random.lognormal(mean=mu, sigma=sigma, size=size)

lung_data = generate_simulated_data(11)
brain_data = generate_simulated_data(5)
bone_data = generate_simulated_data(5)
liver_data = generate_simulated_data(6)

# Create DataFrame for seaborn
data = []
for site, survival_data in zip(sites, [lung_data, brain_data, bone_data, liver_data]):
    for value in survival_data:
        data.append({'Metastatic Site': site, 'Overall Survival (Months)': value})

df = pd.DataFrame(data)

# Create violin plot
plt.figure(figsize=(12, 7))
ax = sns.violinplot(x='Metastatic Site', y='Overall Survival (Months)',
                   data=df, palette=['skyblue', 'lightgreen', 'salmon', 'gold'],
                   inner='quartile', cut=0)

# Add median markers
for i, med in enumerate(median_survival):
    plt.scatter(i, med, color='white', s=80, zorder=3, edgecolor='black')
    plt.text(i, med+1, f'Median: {med}', ha='center', fontsize=10)

# Format plot
plt.title('Survival Distribution by Metastatic Site in Lung Adenocarcinoma', fontsize=14)
plt.xlabel('Site of Single Metastasis', fontsize=12)
plt.ylabel('Overall Survival (Months)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add note about simulated distribution based on medians
plt.text(0.5, -0.12,
         'Note: Distributions are simulated based on reported median values',
         transform=plt.gca().transAxes, ha='center', fontsize=10, style='italic')

# Add citation
plt.text(0.5, -0.17, 'Citation: [1] pmc.ncbi.nlm.nih.gov/articles/PMC10082283',
         transform=plt.gca().transAxes, ha='center', fontsize=8)

plt.ylim(0, 30)  # Set reasonable y-axis limit

plt.tight_layout()
plt.show()
```
