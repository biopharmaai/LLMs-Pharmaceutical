```python
import matplotlib.pyplot as plt
import numpy as np
from lifelines import KaplanMeierFitter

# Create time points (in years) - 5-year follow-up data
time_points = np.array([0, 1, 2, 3, 4, 5])

# PFS rates based on CROWN study comparing lorlatinib vs crizotinib in ALK-rearranged NSCLC
# Source: mdpi.com/2072-6694/17/3/353
# "the 5-year PFS rate was 60% with lorlatinib versus 8% with crizotinib"
# Exact interim points not provided in the data, so approximating based on typical decay patterns
lorlatinib_pfs = np.array([1.0, 0.85, 0.76, 0.70, 0.65, 0.60])  # 60% at 5 years
crizotinib_pfs = np.array([1.0, 0.60, 0.35, 0.20, 0.12, 0.08])  # 8% at 5 years

# Create KaplanMeierFitter objects
kmf_lorlatinib = KaplanMeierFitter()
kmf_crizotinib = KaplanMeierFitter()

# Fit the data
kmf_lorlatinib.fit(time_points, lorlatinib_pfs, label="Lorlatinib")
kmf_crizotinib.fit(time_points, crizotinib_pfs, label="Crizotinib")

# Create plot
plt.figure(figsize=(10, 6))
ax = plt.subplot(111)

# Plot the curves
kmf_lorlatinib.plot_survival_function(ax=ax, color='blue')
kmf_crizotinib.plot_survival_function(ax=ax, color='red')

# Add text annotation about the HR
plt.annotate("HR: 0.28, p < 0.001", xy=(3, 0.4), xytext=(3, 0.4),
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

# Customize plot
plt.title('Progression-Free Survival: Lorlatinib vs Crizotinib in ALK+ NSCLC')
plt.xlabel('Time (Years)')
plt.ylabel('Progression-Free Survival Probability')
plt.ylim(0, 1.05)
plt.grid(True, linestyle='--', alpha=0.7)

# Add text citation
plt.figtext(0.5, 0.01, "Data source: mdpi.com/2072-6694/17/3/353", ha="center", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Data from the report
# Source information
sources = {
    "EGFR": "mdpi.com/2072-6694/17/3/353",
    "KRAS G12C": "mdpi.com/2072-6694/17/3/353",
    "ROS1": "mdpi.com/2072-6694/17/3/353",
    "RET": "mdpi.com/2075-4426/13/2/167",
    "MET": "mdpi.com/2072-6694/17/3/353",
    "HER2": "mdpi.com/2072-6694/17/3/353"
}

# Create data structure for response rates by molecular subtype
data = []

# EGFR mutations with osimertinib+chemo vs osimertinib
# "the combination therapy showed a greater median PFS of 23.7 months vs. 16.6 months with osimertinib monotherapy"
data.append({"Molecular_Subtype": "EGFR", "Treatment": "Osimertinib", "Months_PFS": 16.6})
data.append({"Molecular_Subtype": "EGFR", "Treatment": "Osimertinib+Chemo", "Months_PFS": 23.7})

# KRAS G12C mutations with sotorasib
# "The median PFS was 6.8 months (95% CI, 5.1–8.2)"
data.append({"Molecular_Subtype": "KRAS G12C", "Treatment": "Sotorasib", "Months_PFS": 6.8})

# ROS1 Fusion
# "repotrectinib... a PFS of 34.1 months (95% CI, 25.6-NE) in TKI-naïve patients versus... a PFS of 9.0 months (95% CI, 6.8–19.6) in previously treated patients"
data.append({"Molecular_Subtype": "ROS1", "Treatment": "Repotrectinib (Naive)", "Months_PFS": 34.1})
data.append({"Molecular_Subtype": "ROS1", "Treatment": "Repotrectinib (Pretreated)", "Months_PFS": 9.0})

# RET Fusion
# "median PFS was ~16.5 months in pretreated RET-fusion patients"
data.append({"Molecular_Subtype": "RET", "Treatment": "Selpercatinib", "Months_PFS": 16.5})

# MET Exon 14 Skipping
# "the median PFS was 8.9 months" with tepotinib
# "The mPFS was 5.4 months [after prior therapy] and 12.4 months [when given first-line]"
data.append({"Molecular_Subtype": "MET", "Treatment": "Tepotinib", "Months_PFS": 8.9})
data.append({"Molecular_Subtype": "MET", "Treatment": "Capmatinib (Naive)", "Months_PFS": 12.4})
data.append({"Molecular_Subtype": "MET", "Treatment": "Capmatinib (Pretreated)", "Months_PFS": 5.4})

# HER2 Mutations
# "T-DXd achieved a 55% ORR, a median PFS of 8.2 months"
data.append({"Molecular_Subtype": "HER2", "Treatment": "T-DXd", "Months_PFS": 8.2})
# "pyrotinib produced an ORR of 19.2% (95% CI, 11.2–30.0), with median PFS of 5.6 months"
data.append({"Molecular_Subtype": "HER2", "Treatment": "Pyrotinib", "Months_PFS": 5.6})

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the violin plot
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Create violin plot
ax = sns.violinplot(x="Molecular_Subtype", y="Months_PFS", data=df,
                    inner="points", palette="Set2")

# Add individual data points
sns.stripplot(x="Molecular_Subtype", y="Months_PFS", data=df,
              size=8, jitter=True, edgecolor="gray", linewidth=1,
              hue="Treatment", palette="deep")

# Customize the plot
plt.title("Progression-Free Survival by Molecular Subtype in Lung Adenocarcinoma")
plt.xlabel("Molecular Subtype")
plt.ylabel("Progression-Free Survival (Months)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Treatment", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add data sources as citation
citation_text = "Data sources: EGFR, KRAS G12C, ROS1, MET, HER2: mdpi.com/2072-6694/17/3/353; RET: mdpi.com/2075-4426/13/2/167"
plt.figtext(0.5, 0.01, citation_text, ha="center", fontsize=8)

plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```
