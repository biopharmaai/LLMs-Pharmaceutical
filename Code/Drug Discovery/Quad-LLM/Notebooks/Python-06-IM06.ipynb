```python
# 10. Alternative visualization: GRADE Evidence Quality Assessment
# Note: Stacked bar chart of response categories couldn't be created due to lack of specific response data
# Data source: Summary of Findings table in the meta-analysis

import matplotlib.pyplot as plt
import numpy as np

# GRADE assessment data from the meta-analysis
outcomes = [
    'Overall Survival (OS)',
    'Progression-Free Survival (PFS)',
    '1-year survival',
    '6-month PFS (PFS-6)',
    'Grade ≥3 adverse events',
    'Quality of Life'
]

# GRADE certainty ratings (4=High, 3=Moderate, 2=Low, 1=Very Low)
certainty_ratings = [3, 2, 3, 2, 3, 3]  # From Summary of Findings table

# Key findings
findings = [
    'HR 0.50 (0.33-0.75)',
    'HR 0.85 (0.58-1.22), ns',
    '~28% vs ~20%',
    '~13% vs 10%, ns',
    '56% vs 40%',
    'No significant difference'
]

# Reasons for downgrading
downgrade_reasons = [
    'Open-label phase II',
    'Imprecision & limitations',
    'Open-label phase II',
    'Imprecision & limitations',
    'Open-label phase II',
    'Open-label, not primary'
]

# Create figure
fig, ax = plt.subplots(figsize=(12, 8))

# Set bar positions
y_pos = np.arange(len(outcomes))
bar_heights = [rating for rating in certainty_ratings]

# Create horizontal bars with colors based on certainty
colors = ['red', 'orange', 'yellow', 'green']
bar_colors = [colors[rating-1] for rating in certainty_ratings]

# Plot bars
bars = ax.barh(y_pos, bar_heights, align='center', color=bar_colors, alpha=0.7, edgecolor='black')

# Add outcome labels
for i, outcome in enumerate(outcomes):
    ax.text(-0.5, i, outcome, ha='right', va='center', fontsize=10, fontweight='bold')

# Add findings to the right of bars
for i, finding in enumerate(findings):
    ax.text(certainty_ratings[i] + 0.1, i, finding, va='center', fontsize=9)

# Add downgrade reasons
for i, reason in enumerate(downgrade_reasons):
    ax.text(certainty_ratings[i]/2, i, f"↓ {reason}", va='center', ha='center', fontsize=8,
            color='white', fontweight='bold')

# Customize axis
ax.set_yticks([])
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['Very Low', 'Low', 'Moderate', 'High'], fontsize=10)
ax.set_xlabel('Certainty of Evidence (GRADE)', fontsize=12)
ax.set_title('GRADE Assessment: Quality of Evidence\nRegorafenib vs Lomustine in First-Recurrence Glioblastoma', fontsize=14)
ax.set_xlim(0, 5)
ax.invert_yaxis()  # To have first outcome at the top

# Add a vertical line at each certainty level
for i in range(1, 5):
    ax.axvline(x=i, color='gray', linestyle='--', alpha=0.3)

# Add annotation about evidence base
study_info = (
    "Evidence base: 1 randomized trial (REGOMA, N=119)\n"
    "• Primary endpoint: Overall Survival\n"
    "• Secondary endpoints: PFS, safety, quality of life\n"
    "• Study limitations: Open-label design, baseline imbalances"
)
plt.figtext(0.5, 0.02, study_info, ha='center', fontsize=10,
           bbox=dict(facecolor='lightgray', alpha=0.5, boxstyle='round,pad=0.5'))

plt.tight_layout()
plt.subplots_adjust(left=0.25, bottom=0.15)
plt.show()
```
